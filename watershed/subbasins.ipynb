```python
# Drop columns we won't use for analyses.
# TODO: double-check UNITS matches SWAT.
df_obs = df_obs.drop(['VALUE_QUALIFIER','REMARK_CODE1', 'REMARK_CODE', 'REMARK_CODE', 'ANAL_METH', 'UNITS'], axis=1)
```

```python
# Remove the extra timestamp at the end of the date.
df_obs['DATE'] = df_obs['DATE'].apply(lambda s: s.split(' ')[0])
```

```python
#df_obs
```

```python
obs_nnotur = df_obs[df_obs['PARAMETER'] == 'NITRATES TOTAL,   UNFIL.REAC']
#obs_nnotur
```

```python
# Looks like we only have 1994-1999.
d_obs = {}
for i,r in obs_nnotur.iterrows():
    try:
        year = r['DATE'].split('-')[0]
    except:
        print(year)
    if year in d_obs:
        d_obs[year].append(r['RESULT_VALUE'])
    else:
        d_obs[year] = [r['RESULT_VALUE']]
print d_obs
```

```python
#df['NO3_SURQ']
```

```python
# 10 measures ('NO') per year.
# Loop by subbasins.
d = {}
for i in range(1,32):
    #dx = {yr:[] for yr in dp['YEAR'].unique()}
    dx = {}
    # Isolate the years we have observations for && loop by subbasin.
    dp = df[['NO','SUB','YEAR','FOREST_COVER', 'PCP','NO3_SURQ']][(df['SUB'] == i) & (df['YEAR'] <= 1999) & (1994 <= df['YEAR'])]
    #print(dp)
    for year in dp['YEAR'].unique():
        # Group different "NO"s together by year.
        ds = dp[(dp['YEAR']==year) & (dp['SUB']==i)]
        #print(ds)
        #print('cat')
        if ds.empty:
            print('ERROR: Year {0}, Basin {1}'.format(year,i))
            print(dp)
        dx[year] = ds
    d[i] = dx
```

```python
import numpy as np
from scipy import stats

# So forest cover doesn't really change between years.
# And some basin share values.
for basin in d:
    predictors = d[basin]
    forest = []
    pcp = []
    #print(predictors)
    for year in predictors:
        x = predictors[year]
        # This is probably a bad idea.
        #print(basin,year,np.mean(x['FOREST_COVER']), np.mean(x['PCP']))
        forest.append(np.mean(x['FOREST_COVER']))
        pcp.append(np.mean(x['PCP']))
    #print(forest)
    #print(pcp)
    predictands = [np.mean(l) for l in d_obs.itervalues()]
    #print(predictands)
    f_rho, f_pval = stats.spearmanr(forest,predictands)
    p_rho, p_pval = stats.spearmanr(pcp,predictands)
    print('f_rho: {0}, f_pval: {1}, p_rho: {2}, p_val: {3}, basin: {4}'.format(f_rho,f_pval,p_rho,p_pval,basin))
```

```python
# Let's create a linear model based of basin 10.
predictors = d[10]
forest = []
pcp = []
no3_surq = []
for year in predictors:
    x = predictors[year]
    # This is probably a bad idea, but we need 1 value per year.
    forest.append(np.mean(x['FOREST_COVER']))
    pcp.append(np.mean(x['PCP']))
    no3_surq.append(np.mean(x['NO3_SURQ']))
```

```python
df = pd.DataFrame({'forest':forest,'pcp':pcp})
```

```python
df
```

```python
from sklearn.cross_validation import KFold
from sklearn import linear_model

# Now let's compute RMSE using 5-fold x-validation
kf = KFold(len(no3_surq), n_folds=5)
xval_err = 0
for train,test in kf:
    #print(train,test)
    X = df[df.index.isin(train)]
    y = [no3_surq[i] for i in train]
    # Create linear regression object# Create
    lm = linear_model.LinearRegression()
    model = lm.fit(X,y)
    p = model.predict(df[df.index.isin(test)])
    e = p-[no3_surq[i] for i in test]
    #print(p,e)
    xval_err += np.dot(e,e)
    
rmse_5cv = np.sqrt(xval_err/len(no3_surq))
```

```python
method_name = 'Simple Linear Regression'
print('Method: %s' %method_name)
print('RMSE on 5-fold CV: %.4f' %rmse_5cv)
```
